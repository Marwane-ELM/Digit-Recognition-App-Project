```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, KFold, learning_curve, GridSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from skopt import BayesSearchCV

import os
from joblib import dump, load
```

## Import du dataset

```python
mnist = np.load("../data/mnist_prepared.npz")
```

```python
X = mnist["X"]
Y = mnist["Y"]
```

### Train_test_split

```python
def persistent_train_test_split(x, y, path, train_size):

    t_size = str(train_size).replace('.', '_')
    file_name = "GaussNB_data_split" + '_' + t_size
    path_train_size = path + file_name
    
    if os.path.isfile(path_train_size):
        X_train, X_test, Y_train, Y_test = load(path_train_size)
        print("Le split a été importé avec succès")
    else :         
        X_train, X_test, Y_train, Y_test = train_test_split(
            x,
            y,
            train_size = train_size,   # On entraine sur 5% de notre dataset
            stratify = y,
            shuffle = True,
            random_state = 42
        )
        
        dump((X_train, X_test, Y_train, Y_test), path_train_size)
        print("Le split a été créé et importé avec succès")

    return X_train, X_test, Y_train, Y_test
```

```python
GNB_path_data = "../data/"

X_train, X_test, Y_train, Y_test = persistent_train_test_split(X, Y, GNB_path_data, 0.05)
```

```python
Y_train.shape
```

## Gaussian Naive Bayes model

##### Fonction de sauvegarde de modèles entraînés

```python
def dump_artefact(tuned_model, path_artefact : str, artefact_name : str):

    
    score = round(tuned_model.best_score_, 3)  # on arrondis le score à 3 décimales
    score = str(score).replace('.', '')
    name = artefact_name + '_' + score

    path = path_artefact + name
    if os.path.isfile(path):
        print(f"Le modèle {artefact_name} existe déjà")

    else :
        model = dump(tuned_model.best_estimator_, path)  # Si le modèle a déjà été crée précédemment, alors on l'importe
        print(f"L'artefact de {artefact_name} a été crée avec succès")
        
```

#### Pipeline

```python
Pipeline_KNC = Pipeline([
    ('std_scaler', StandardScaler()),
    ('knc', KNeighborsClassifier(weights="distance", algorithm="auto", metric="minkowski"))
])
```

#### Recherche d'hyperparamètres n°1

```python
param_grid_KNC1 = {
        "knc__p" : [1, 2, 3],
        "knc__n_neighbors" : np.arange(1, 23, 2)
}
```

```python
Grid_KNC1 = BayesSearchCV(
    estimator = Pipeline_KNC,
    search_spaces = param_grid_KNC1,
    n_iter = 20,
    scoring = 'accuracy',
    cv = KFold(n_splits = 3),
    verbose = 4,
    n_jobs = 13
)
```

```python
Grid_KNC1.fit(X_train, Y_train)
```

```python
print("Best params : ", Grid_KNC1.best_params_ )
print("Best score : ", Grid_KNC1.best_score_)
```

```python
dump_artefact(
    tuned_model = Grid_KNC1,
    path_artefact = "../saved_models/KNC_persistent/",
    artefact_name = "KNC1"
)
```

### Analyse du meilleur modèle

```python
def best_model(path):
    biggest_score = 0
    best_model = ""
    for artefact_name in os.listdir(path):
        if artefact_name != ".ipynb_checkpoints":
            score = int(artefact_name[-4:])
            if score > biggest_score:
                biggest_score = score
                best_model = artefact_name
    path_best_model = path + best_model
    model = load(path_best_model)
    return model
            
```

```python
Best_model = best_model("../saved_models/KNC_persistent/")
```

#### Comparaison des scores entre le train_set et le test_set issus du train_test_split n°3

```python
print("Train_score : ", Best_model.score(X_train3, Y_train3) )
print("Test_score : ", Best_model.score(X_test3, Y_test3) )
```

#### Learning Curve

```python
N, train_scores, val_scores = learning_curve(Best_model, X_train3, Y_train3, cv=3, train_sizes=np.linspace(0.1, 1, 12))
```

```python
plt.figure(figsize=(5, 4))
plt.plot(N, train_scores.mean(axis=1), c='blue', label="train_score")
plt.plot(N, val_scores.mean(axis=1), c='orange', label="val_score")
plt.legend()
```

### Conclusion

Le modèle n'est pas en underfitting car on remarque qu'il apprend très bien, voire même trop. On peut constater cela avec le score sur le train set qui a une accuracy de 100%.  
On pourrait penser qu'on est en underfitting mais ce n'est pas le cas ici car c'est la nature de notre dataset qui offre un score si élevé.  
Les images sont des vecteurs à 784 dimensions donc les différentes classes d'image sont bien éloignées entre elles dans l'espace. Deux chiffres identiques ont une faible distance dans l'espace.  
Par précaution, on va quand même utiliser 5 voisins pour mieux généraliser et éviter les cas où une classe ne se démarque lors des votes durant les prédictions.

```python
final_model = KNeighborsClassifier(
    n_neighbors=5,
    weights="distance", 
    algorithm="auto", 
    metric="euclidean"
)

final_model.fit()
```

```python
if os.path.isfile("../saved_models/KNC_persistent/KNC_1000"):
        print("Le modèle existe déjà")
else :
    model = dump(final_model, "../saved_models/KNC_persistent/KNC_1000")  # Si le modèle a déjà été crée précédemment, alors on l'importe
    print("L'artefact été crée avec succès")
```
